```python
# change dat file to csv
import re
import csv
from pathlib import Path

# Change these
DAYS = [
    "05-01-2026",
    "06-01-2026",
    "07-01-2026",
]
infile = Path("00.05.dat")     # your .dat file
outfile = Path("00.05.csv")

float_re = re.compile(r"""^[\s+-]?(?:\d+\.?\d*|\.\d+)(?:[eE][+-]?\d+)?$""")

def is_number(tok: str) -> bool:
    return bool(float_re.match(tok))

rows = []
with infile.open("r", errors="ignore") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        # Split on whitespace (works for tabs/spaces)
        parts = line.split()

        # Your sample looks like 5 numeric columns.
        if len(parts) == 5 and all(is_number(p) for p in parts):
            rows.append([float(p) for p in parts])

# Write CSV
with outfile.open("w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["analog", "analog_sterr", "photon_counting", "pc_sterr", "overflow_info"])
    w.writerows(rows)

print(f"Saved {len(rows)} rows -> {outfile}")
```
